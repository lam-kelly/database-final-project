```python
import pandas as pd
import numpy as np
df = pd.read_csv("customer-churn.csv")
# def introduce_nan(x,percentage):
#     n = int(len(x)*(percentage - x.isna().mean()))
#     idxs = np.random.choice(len(x), max(n,0), replace=False, p=x.notna()/x.notna().sum())
#     x.iloc[idxs] = np.nan
# df.apply(introduce_nan, percentage=.1)
df.loc[df.sample(frac=0.1).index, "Age"] = np.nan
df.info()
```

```python
# Separate the null values from dataframe (df) and create a variable “test data”
test_data = df[df['Age'].isnull()]

# Drop the null values from the dataframe (df) and represent as ‘train data”
df.dropna(inplace=True)
```

```python
df
```

```python
# Create “x_train” & “y_train” from train data.
x_train = df.drop(['Age'], axis=1)
x_train
```

```python
y_train = df['Age']
y_train
```

```python
# Build the linear regression model
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)

x_test = test_data[['Call  Failure','Complains','Subscription  Length','Charge  Amount','Seconds of Use','Frequency of use','Frequency of SMS','Distinct Called Numbers','Age Group','Tariff Plan','Status','Customer Value','FN','FP','Churn']]
```

```python
# Apply the model on x_test of test data to make predictions.
y_pred = lr.predict(x_test)

# Replace the missing values with predicted values.
test_data['Age'] = y_pred
```

```python
test_data
```

```python
actual = pd.read_csv("customer-churn.csv")
final = actual[actual.index.isin(test_data.index)]
comp = final['Age'].sub(test_data['Age'])
comp
```

```python
from sklearn.metrics import mean_squared_error
rms = mean_squared_error(test_data["Age"], final["Age"])
rms
```
