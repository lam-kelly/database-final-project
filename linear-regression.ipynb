```python
import pandas as pd
import numpy as np
sample_customer_data = pd.read_csv("customer-churn.csv",  nrows=100)
# sample_customer_data.info()
def introduce_nan(x,percentage):
    n = int(len(x)*(percentage - x.isna().mean()))
    idxs = np.random.choice(len(x), max(n,0), replace=False, p=x.notna()/x.notna().sum())
    x.iloc[idxs] = np.nan
sample_customer_data.apply(introduce_nan, percentage=.5)
sample_customer_data.info()
```

```python
mean_value = sample_customer_data.mean()
mean_imputation = sample_customer_data.fillna(mean_value)
mean_imputation.head()
```

```python
sample_customer_data = pd.read_csv("customer-churn.csv",  nrows=100)
sample_customer_data.compare(mean_imputation)
```

```python
import pandas as pd

df = pd.read_csv('mbta.csv')
# df.info()
# df.shape
df.isnull().sum()
```

```python
# Separate the null values from dataframe (df) and create a variable “test data”
test_data = df[df['total_ons'].isnull()]

# Drop the null values from the dataframe (df) and represent as ‘train data”
df.dropna(inplace=True)
```

```python
# Create “x_train” & “y_train” from train data.
x_train = df.drop(['line_id','time_period_id','station_id','total_ons'], axis=1)
y_train = df['total_ons']
```

```python
# y_train
```

```python
# Build the linear regression model
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)

x_test = test_data[['direction']]
```

```python
# Apply the model on x_test of test data to make predictions.
y_pred = lr.predict(x_test)

# Replace the missing values with predicted values.
test_data['y_pred'] = y_pred
```
